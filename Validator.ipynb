```python
import logging as l
import json
```

```python
json_filename = "trader_joes.json"
```

```python
class Validator:
    def __init__(self, output):
        self.output = output
        
    def validate_price(self):
        
        for data in output:
            if 'sale_price' in data.keys() and data['sale_price'] <= data['price'] and data['sale_price'] != 0:
                pass
            else:
                l.error("Please add/check price for product_id {}".format(data['product_id']))
    
    def validate_fields(self):
        for data in output:
            if 'title' in data.keys() and data['title'] != '':
                pass
            else:
                l.error("Title is mandatory. Please add title for product_id {}".format(data['product_id']))
            
            if 'url' in data.keys() and data['title'] != '':
                pass
            else:
                l.error("Product url is mandatory. Please add url for product_id {}".format(data['product_id']))
                
            if 'product_id' in data.keys() and data['product_id'] != '':
                pass
            else:
                l.error("Product id is mandatory. Please add product id in data.")
                
            if 'models' in data.keys() and 'variants' in data['models'].keys() and data['models']['variants'] != []:
                pass
            else:
                l.error("Variants are mandatory. Please add variants for product_id {}.".format(data['product_id']))
            
    def validate_variants(self):
        for data in output:
            if 'variants' in data['models'].keys() and data['models']['variants'] != []:
                for variant in data['models']['variants']:
                    if variant['id'] != '':
                        pass
                    else:
                        l.error("Variant id is mandatory. Please add variant id for product_id {}.".format(data['product_id']))
                        
                    if variant['image'] != '':
                        pass
                    else:
                        l.error("Variant image is not available. Please add variant image for product_id {}.".format(data['product_id']))
                    
                    if variant['price'] != '':
                        pass
                    else:
                        l.error("Variant price is not available. Please add variant price for product_id {}.".format(data['product_id']))  
                    
                    if variant['size'] != '':
                        pass
                    else:
                        l.error("Variant size is not available. Please add variant size for product_id {}.".format(data['product_id']))

                    if variant['color'] != '':
                        pass
                    else:
                        l.error("Variant color is not available. Please add variant color for product_id {}.".format(data['product_id']))
            else:
                l.error("Variants are not available for product_id {}".format(data['product_id']))
```

```python
with open("output/"+json_filename, "r") as json_file:
  output = json.load(json_file)
```

```python
validator = Validator(output)
```

```python
validator.validate_price()
```

```python
validator.validate_fields()
```

```python
 validator.validate_variants()   
```
