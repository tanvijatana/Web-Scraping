```python
import requests
from bs4 import BeautifulSoup
import json
import logging as l
```

```python
url_list = ["https://www.lechocolat-alainducasse.com/uk/easter-chocolate",
           "https://www.lechocolat-alainducasse.com/uk/chocolates",
           "https://www.lechocolat-alainducasse.com/uk/chocolate-bar",
           "https://www.lechocolat-alainducasse.com/uk/chocolate-gift",
           "https://www.lechocolat-alainducasse.com/uk/simple-pleasures",
           "https://www.lechocolat-alainducasse.com/uk/breakfast-snacks"]
```

```python
output = []

for url in url_list:
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'lxml')
    brand = soup.select_one("title").get_text()
    description = soup.select_one("meta[name='description']")['content']
    products_container = soup.select("section[id='js-product-list'] div[class='productMiniature js-product-miniature']")
    for con in products_container:
        temp = {}
        temp['brand'] = brand
        temp['description'] = description
        temp['image'] = con.select_one("header div figure img")['src']
        temp['price'] = con.select_one("footer div span").get_text().replace('\n','').replace('£','').strip()
        temp['sale_price'] = con.select_one("footer div span").get_text().replace('\n','').replace('£','').strip()
        temp['title'] = con['data-product-name']
        temp['url'] = con.select_one("section a")['href']
        if con.select_one("section a h3 span"):
            temp['quantity'] = con.select_one("section a h3 span").get_text().replace('\n','').strip()
        else:
            temp['quantity'] = ''
        temp['product_id'] = con['data-id-product']
        output.append(temp)
```
